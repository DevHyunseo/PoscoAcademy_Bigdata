#### 유기농 물품 변수 생성

```python
# 유기농 제품의 정보를 확인하고 분류
pattern = r'/유기|유기농|유|논지엠오|무농약|무항생제'
dfm['유기농물품'] = dfm['물품명'].str.contains(pattern, regex=True)

# '유기농 물품' 열을 범주형 'O'와 'X'로 변환
dfm['유기농 물품'] = dfm['유기농물품'].apply(lambda x: 'O' if x else 'X')
del dfm['유기농물품']

# 수정된 데이터프레임을 출력
dfm.head()
```

#### 연령대 변수 생성

```python
# 연령을 10년 단위로 나이대 분류하여 연령대 변수 생성
bins = [0, 29, 39, 49, 59, 69, float('inf')]
labels = ['20대 이하', '30대', '40대', '50대', '60대', '70대 이상']
df['연령대'] = pd.cut(df['연령'], bins=bins, labels=labels, right=False)
df.head()
```

#### 지역 변수 생성

```python
# 구 분류 기준 정의
# (1) '수지'를 제외한 '용인시'와 인접한 '수원시', '성남시'의 지역을 주변지역으로 분류
# (2) 그 외의 지역들을 '먼 지역'으로 분류

def classify_region(gu):
    suji_region = ['수지구']
    nearby_region = ['기흥구', '분당구', '수원 영통구', '수원 팔달구', '수원 장안구', '수원 권선구',  '처인구','중원구', '수정구']
    distant_region = ['서울 강남구', '서울 송파구', '화성', '광주', '하남','기타']

    if gu in suji_region:
        return '수지 지역'
    elif gu in nearby_region:
        return '주변 지역'
    elif gu in distant_region:
        return '먼 지역'
    else:
        return '기타'

# 지역 칼럼 추가
df['지역'] = df['구'].apply(classify_region)
```
