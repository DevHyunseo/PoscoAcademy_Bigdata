```python
import pandas as pd
```

```python
df = pd.read_csv("/home/piai/바탕화면/빅데이터 프로젝트/product_data.csv",encoding = "euc_kr")
df
```

```python
df
```

```python
# 회원번호를 인덱스로 설정하고, 각 구입 품목을 칼럼으로 변환하는 피벗 테이블 생성
pivot_df = df.pivot_table(
    index='회원번호',
    columns='물품소분류',
    values='구매수량',
    aggfunc='sum',
    fill_value=0)
pivot_df_TF = pivot_df > 0
pivot_df_TF
```

```python
# ▶ max_len =2, 최대 2개까지의 리스트가 들어간 연관규칙 탐색

from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemset = apriori(pivot_df_TF,
                           min_support=0.05,
                           max_len=2,
                           use_colnames=True)

frequent_itemset['length'] = frequent_itemset['itemsets'].apply(lambda x: len(x))
frequent_itemset.sort_values('support',ascending=False,inplace=True)
```

```python
# ▶ 상위에서 생성한 Rule에 대해, 추가 지표들에 대한 탐색
# ▶ metric과 threshold 설정 필요
association_rules_df = association_rules(frequent_itemset,
                                         metric='confidence',
                                         min_threshold=0.005)
```

```python
# 공급일자와 구매시각을 합쳐서 구매일시 컬럼 생성
df['구매일시'] = df['공급일자'] + ' ' + df['구매시각']

# 결과 확인
df.head()
```

```python
# 회원번호를 인덱스로 설정하고, 각 구입 품목을 칼럼으로 변환하는 피벗 테이블 생성
pivot_df = df.pivot_table(
    index=['회원번호','구매일시'],
    columns='물품소분류',
    values='구매수량',
    aggfunc='sum',
    fill_value=0)
pivot_df_TF = pivot_df > 0
pivot_df_TF
```

## 2) 연관분석

```python
pivot_df_TF.isnull().sum().sum()
```

```python
# ▶ max_len =2, 최대 2개까지의 리스트가 들어간 연관규칙 탐색

from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemset = apriori(pivot_df_TF,
                           min_support=0.001,
                           max_len=2,
                           use_colnames=True)

frequent_itemset['length'] = frequent_itemset['itemsets'].apply(lambda x: len(x))
frequent_itemset.sort_values('support',ascending=False,inplace=True)
```

```python
# ▶ 전체 연관 규칙 출력
frequent_itemset
```

```python
frequent_itemset['length'].unique()
```

```python
frequent_itemset[frequent_itemset['length'] >= 2].head(10)
```

```python
# ▶ 특정 규칙을 찾고 싶을 때, 정확한 규칙을 넣어줘야 함
frequent_itemset[frequent_itemset['itemsets'] == frozenset(('콩나물','두부'))]
```

```python
# ▶ 상위에서 생성한 Rule에 대해, 추가 지표들에 대한 탐색
# ▶ metric과 threshold 설정 필요
association_rules_df = association_rules(frequent_itemset,
                                         metric='confidence',
                                         min_threshold=0.005)
```

```python
association_rules_df
```

- confidence는 조건부확률 A구매했을 때, B를 구매할 확률
- Lift는 랜덤으로 콩나물을 살 확률이 분모 콩나물이랑 두부를 함께 구매할 확률이 분자 / 1이상이면 두개는 관련이 있다.
- 한마디로 "자연스럽게 구매하는 확률보다 높은가?"

```python
# ▶ 특정 규칙을 찾고 싶을 때
association_rules_df[association_rules_df['consequents']==frozenset({'두부'})]
# 선행구매, 후행구매 나누어서 확인가능
# 여기서는 두부를 뒤에 구매한 경우만
```

### ·유의미한 규칙 탐색

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemset = apriori(pivot_df_TF,
                           min_support=0.001,
                           max_len=2,
                           use_colnames=True)

frequent_itemset['length'] = frequent_itemset['itemsets'].map(lambda x: len(x))
frequent_itemset.sort_values('support',ascending=False,inplace=True)
```

```python
frequent_itemset
```

```python
# rules['support'].mean(), rules['confidence'].mean() #이러면 안돼..
```

```python
# ▶ lift가 1이상인 유의미한 규칙만 탐색 (랜덤 적인 효과 대비 우수) 이게 제일 중요
rules = association_rules(frequent_itemset, metric="lift", min_threshold=1).sort_values(by = ['lift',  'confidence','support'], ascending =False)
rules
#support, lift 순으로 해야할 때도 있을 support가 많이 낮다 생각될 때
# 총 유의미한 관계는 54개 중복존재
```

```python
# ▶ 많이 발견되는 규칙보고 싶을 때 (※ Support)
rules = association_rules(frequent_itemset, metric="support", min_threshold=0.05).sort_values(by = ['support', 'lift', 'confidence'], ascending =False)
rules
```
