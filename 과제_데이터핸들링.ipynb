```python
#실습4 : 정규분포에서 추출한 10개의 무작위 데이터로 데이터 선택하기

from numpy.random import randn

#step1 : 데이터 구상하기
np.random.seed(1234) #고정된 시드 :1234
df = pd.DataFrame(randn(10,4), columns=['A', 'B', 'C', 'D'])

#step2 : slice
#(1)B열 선택
df[['B']]

#(2)B, D열 선택
df[['B', 'D']]

#(3)4~8행 선택
df[4:9]

#step3 : loc, iloc
#(1) loc : 2~8행, A,C열 선택
df.loc[2:8, ['A', 'C']]

#(2) loc : A,C열 선택
df.loc[:,['A', 'C']]

#(3) loc : 3~8행 선택
df.loc[3:8]

#(4) loc : 3행 선택
df.loc[3:3]

#(5) iloc :0~6행, A열 선택
df.iloc[0:7, [0]]

#(6) iloc : A열 선택
df.iloc[:, [0]]

#(7) iloc : 2~6행 선택
df.iloc[2:7]

#(8) iloc : 5행 선택
df.iloc[5:6]

#(9) query : d > 0.5 선택
df.query('D > 0.5')

#(10) query : B > D 선택
df.query('B > D')
df[df.B > df.D]

#(11) query : A > 0.5 그리고 D < 0.0 선택
df.query('A > 0.5 & D < 0.0')

#(12) query : A > 0.5 혹은 D < 0.0 선택
df.query('A > 0.5 | D < 0.0')
```

```python
#실습5 : df4, df5, df6 concat하기

#step1 : 데이터 구성하기
df4 = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'],
                  'B':['B0', 'B1', 'B2', 'B3'],
                  'C':['C0', 'C1', 'C2', 'C3'],
                  'D':['D0', 'D1', 'D2', 'D3']})
df5 = pd.DataFrame({'E':['E0', 'E1', 'E2', 'E3'],
                  'F':['F0', 'F1', 'F2', 'F3'],
                  'G':['G0', 'G1', 'G2', 'G3'],
                  'H':['H0', 'H1', 'H2', 'H3']})
df6 = pd.DataFrame({'I':['I0', 'I1', 'I2', 'I3'],
                  'J':['J0', 'J1', 'J2', 'J3'],
                  'K':['K0', 'K1', 'K2', 'K3'],
                  'L':['L0', 'L1', 'L2', 'L3']})

# #step2 : concat을 이용한 데이터 결합
result = pd.concat([df4, df5, df6])
result

#step3 : 파라미터 변경(axis = 1로 두어서 좌우로 합치기)
result = pd.concat([df4, df5, df6], axis=1)
result
```

```python
#실습7 : pivot을 이용하여 GENDER와 AGEGROUP별 WEIGHT의 평균 나타내기

#step1 : 데이터 구성하기
df = pd.read_csv("/home/piai/바탕화면/빅데이터주간/01 실습파일/1 Big Data 분석 기초/FITNESS.csv",  encoding = 'euc-kr')
df.head()

#step 2: groupby, agg를 이용한 평균값 계산
d1= df.groupby(['GENDER', 'AGEGROUP'], as_index=False).agg({'WEIGHT':'mean'})

#step 3: pivot을 이용한 변환 1
d2 = d1.pivot('GENDER', 'AGEGROUP', 'WEIGHT')

#step3 : pivot을 이용한 변환 2
d3 = d1.pivot('AGEGROUP', 'GENDER', 'WEIGHT')
d3
```
